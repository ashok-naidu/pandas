#lab 37 - Reshaping dataframes in Pandas

# Problem

1. Read 'nba.csv' file using pandas library 

2. Use stack method to change from wide table to 

3. Unstack dataframe 

4. Melt dataframe for columns 'Name' and 'Team'



```python
## Solution

import pandas as pd

df=pd.read_csv("nba.csv")

df.head()
```

```python
# reshape the dataframe using stack() method 

df_stacked = df.stack() 

df_stacked.head(100)
```

```python
# unstack() method 
df_unstacked = df_stacked.unstack() 

df_unstacked.head(10)
```

```python
# take two columns "Name" and "Team" to melt

df_melt = df.melt(id_vars =['Name', 'Team'])  

df_melt.head(10)
```
